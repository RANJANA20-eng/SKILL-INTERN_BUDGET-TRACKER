```python
import tkinter as tk
from tkinter import messagebox

# Initialize variables to store expenses and budgets
expenses = []
budgets = {}

# Function to add an expense
def add_expense():
    expense = expense_entry.get()
    amount = float(amount_entry.get())
    category = category_entry.get()
    
    # Append the expense to the list and update the budget
    expenses.append((expense, amount, category))
    if category in budgets:
        budgets[category] -= amount
    else:
        budgets[category] = -amount
    
    # Clear the input fields
    expense_entry.delete(0, tk.END)
    amount_entry.delete(0, tk.END)
    
    # Show a confirmation message
    messagebox.showinfo("Expense Added", f"Expense '{expense}' added successfully!")

# Function to set a budget
def set_budget():
    category = budget_category_entry.get()
    budget_amount = float(budget_amount_entry.get())
    
    # Update or set the budget for the category
    budgets[category] = budget_amount
    
    # Clear the input fields
    budget_category_entry.delete(0, tk.END)
    budget_amount_entry.delete(0, tk.END)
    
    # Show a confirmation message
    messagebox.showinfo("Budget Set", f"Budget for '{category}' set to ${budget_amount:.2f}")

# Function to view spending habits
def view_spending_habits():
    # Calculate and display spending habits
    spending_habits = {}
    for expense, amount, category in expenses:
        if category in spending_habits:
            spending_habits[category] += amount
        else:
            spending_habits[category] = amount
    
    habits_text = "Spending Habits:\n"
    for category, amount in spending_habits.items():
        habits_text += f"{category}: ${amount:.2f}\n"
    
    habits_label.config(text=habits_text)

# Create the main window
app = tk.Tk()
app.title("Budget Tracking App")

# Create and configure input elements for adding expenses
expense_label = tk.Label(app, text="Expense:")
expense_label.pack()
expense_entry = tk.Entry(app)
expense_entry.pack()

amount_label = tk.Label(app, text="Amount ($):")
amount_label.pack()
amount_entry = tk.Entry(app)
amount_entry.pack()

category_label = tk.Label(app, text="Category:")
category_label.pack()
category_entry = tk.Entry(app)
category_entry.pack()

add_expense_button = tk.Button(app, text="Add Expense", command=add_expense)
add_expense_button.pack()

# Create and configure input elements for setting budgets
budget_category_label = tk.Label(app, text="Budget Category:")
budget_category_label.pack()
budget_category_entry = tk.Entry(app)
budget_category_entry.pack()

budget_amount_label = tk.Label(app, text="Budget Amount ($):")
budget_amount_label.pack()
budget_amount_entry = tk.Entry(app)
budget_amount_entry.pack()

set_budget_button = tk.Button(app, text="Set Budget", command=set_budget)
set_budget_button.pack()

# Create and configure a button to view spending habits
view_habits_button = tk.Button(app, text="View Spending Habits", command=view_spending_habits)
view_habits_button.pack()

# Create and configure a label to display spending habits
habits_label = tk.Label(app, text="", font=("Helvetica", 12))
habits_label.pack()

# Start the Tkinter main loop
app.mainloop()
```
